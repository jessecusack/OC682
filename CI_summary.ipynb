# Confidence Intervals

Confidence intervals relate to the probability that the statistic of interest (mean, variance) is close to the true value which we don't know. Commonly, one reports the 95% confidence interval of a statistic. This does not mean that there is a 95% chance that the interval around the statistic contains the true value. The definition requires a notion of repetition. Specifically, repeated estimates of the statistic bounded by the confidence interval should contain the true mean 95% of the time.

If the parameter of a random variable is $\theta$, we want to identify a range containing $\theta$ such that the estimate of the parameter $\hat{\theta}$ falls within that range with a probability $1 - \alpha$,

\begin{equation}
\text{Prob}[\theta_l < \theta < \theta_u] = 1 - \alpha.
\end{equation}

How do we identify the limits $\theta_l$ and $\theta_u$?

## Confidence Interval on the Mean

The true mean of some data $y$ is $\mu_y$ and the sample estimate of the mean is $\hat{\mu}_y$ and the estimated standard deviation is $\hat{\sigma}_y$

\begin{equation}
\text{Prob}[\hat{\mu}_y + z_{\frac{\alpha}{2}} \hat{\sigma}_y < \mu_y < \hat{\mu}_y + z_{1-\frac{\alpha}{2}} \hat{\sigma}_y] = 1 - \alpha.
\end{equation}

where $z_{1-\frac{\alpha}{2}}$ and $z_{\frac{\alpha}{2}}$ are percent points on the $t$-distribution. The points flank the region containing $1-\alpha$ of the area under the curve. If the number of degrees of freedom in the data are large then the percent points are well approximated using the standard normal distribution. 


```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['xtick.major.top'] = False
plt.rcParams['ytick.major.right'] = False
```

```python
df = 6
alpha = 0.05
z = np.linspace(-4, 4, 200)
tdist = stats.t(df=df)
#norm = stats.norm()

ci_lower, ci_upper = tdist.interval(1-alpha)

fig, ax = plt.subplots()
#ax.plot(z, norm.pdf(z), color='gray', label='Normal')
ax.plot(z, tdist.pdf(z), color='k', label=f't (df={df})')
ax.fill_between(z, tdist.pdf(z), where=(z > ci_lower) & (z < ci_upper), color='lightgray', alpha=0.5)
ax.set_ylim(0, 0.4)

ax.legend()
ax.set_xlabel('z')
ax.set_ylabel('p(z)')
ax.annotate(r'area = 1 - $\alpha$', xy=(0, 0.2), ha='center', va='center', fontsize=12)
ax.annotate(r"$z_{\frac{\alpha}{2}}$", xy=(ci_lower, 0), xytext=(ci_lower, 0.08), ha='center', va='center', fontsize=12, arrowprops=dict(arrowstyle='->', color='k'))
ax.annotate(r"$z_{1-\frac{\alpha}{2}}$", xy=(ci_upper, 0), xytext=(ci_upper, 0.08), ha='center', va='center', fontsize=12, arrowprops=dict(arrowstyle='->', color='k'))
```
